Find unique values from mainroad,guestroom,basement,hotwaterheating ,airconditioning and prefarea :

```python
print(df.mainroad.unique())
print(df.guestroom.unique())
print(df.basement.unique())
print(df.hotwaterheating.unique())
print(df.airconditioning.unique())
print(df.prefarea.unique())
```

Replacing all distorted values as yes and no:


```python
y=['Haanji', 'yes', 'Haan', 'Of course', 'yep', 'y', 'True', 'Y', 'Yo', 'Indeed', "['y']"]
replace_yes='YES'
n=['no', 'Nakko', 'Nope', 'ille', 'N', 'Naaji', 'n', "['n']", 'False', 'Nah', 'Indeed']
replace_no='NO'

df["mainroad"]=df["mainroad"].replace(y,replace_yes)
df["mainroad"]=df["mainroad"].replace(n,replace_no)
df["guestroom"]=df["guestroom"].replace(y,replace_yes)
df["guestroom"]=df["guestroom"].replace(n,replace_no)
df["basement"]=df["basement"].replace(y,replace_yes)
df["basement"]=df["basement"].replace(n,replace_no)
df["hotwaterheating"]=df["hotwaterheating"].replace(y,replace_yes)
df["hotwaterheating"]=df["hotwaterheating"].replace(n,replace_no)
df["airconditioning"]=df["airconditioning"].replace(y,replace_yes)
df["airconditioning"]=df["airconditioning"].replace(n,replace_no)
df["prefarea"]=df["prefarea"].replace(y,replace_yes)
df["prefarea"]=df["prefarea"].replace(n,replace_no)
df
```

Filling all missing values:

```python
df['area'].median()
```

```python
df['area'].fillna(df['area'].median(),inplace=True)
df
```

```python
df['bedrooms'].fillna(df['bedrooms'].median(),inplace=True)
df['bathrooms'].fillna(df['bathrooms'].median(),inplace=True)
df['stories'].fillna(df['stories'].median(),inplace=True)
df['parking'].fillna(df['parking'].median(),inplace=True)
df
```

```python
df['mainroad'].fillna(df['mainroad'].mode()[0],inplace=True)
df['guestroom'].fillna(df['guestroom'].mode()[0],inplace=True)
df['basement'].fillna(df['basement'].mode()[0],inplace=True)
df['hotwaterheating'].fillna(df['hotwaterheating'].mode()[0],inplace=True)
df['airconditioning'].fillna(df['airconditioning'].mode()[0],inplace=True)
df['prefarea'].fillna(df['prefarea'].mode()[0],inplace=True)
df['furnishingstatus'].fillna(df['furnishingstatus'].mode()[0],inplace=True)
df
```
