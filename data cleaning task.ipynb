Replacing all distorted values as yes and no:


```python
y=['Haanji', 'yes', 'Haan', 'Of course', 'yep', 'y', 'True', 'Y', 'Yo', 'Indeed', "['y']"]
replace_yes='YES'
n=['no', 'Nakko', 'Nope', 'ille', 'N', 'Naaji', 'n', "['n']", 'False', 'Nah', 'Indeed']
replace_no='NO'

df["mainroad"]=df["mainroad"].replace(y,replace_yes)
df["mainroad"]=df["mainroad"].replace(n,replace_no)
df["guestroom"]=df["guestroom"].replace(y,replace_yes)
df["guestroom"]=df["guestroom"].replace(n,replace_no)
df["basement"]=df["basement"].replace(y,replace_yes)
df["basement"]=df["basement"].replace(n,replace_no)
df["hotwaterheating"]=df["hotwaterheating"].replace(y,replace_yes)
df["hotwaterheating"]=df["hotwaterheating"].replace(n,replace_no)
df["airconditioning"]=df["airconditioning"].replace(y,replace_yes)
df["airconditioning"]=df["airconditioning"].replace(n,replace_no)
df["prefarea"]=df["prefarea"].replace(y,replace_yes)
df["prefarea"]=df["prefarea"].replace(n,replace_no)
df
```

finding relation between area and price:

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['price'], df['area'])
plt.xlabel('price')
plt.ylabel('area')
plt.show()
```

```python
for i in df.index:
    if df.loc[i,'price'] < 5000000 and np.isnan(df.loc[i, 'area']):
        df.loc[i,'area'] = 4000
    elif df.loc[i,'price'] < 6000000 and np.isnan(df.loc[i, 'area']):
        df.loc[i,'area'] = 5000
    elif df.loc[i,'price'] < 8000000 and np.isnan(df.loc[i, 'area']):
        df.loc[i,'area'] = 6000
    elif df.loc[i,'price'] <10000000 and np.isnan(df.loc[i, 'area']):
        df.loc[i,'area'] = 9000
    elif np.isnan(df.loc[i, 'area']):
        df.loc[i,'area'] = 3000
    else:
        continue

df
```
